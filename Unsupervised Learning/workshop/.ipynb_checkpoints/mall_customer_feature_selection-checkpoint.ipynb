```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
from sklearn.datasets import load_iris, load_diabetes
```

```python
data = pd.read_csv('Mall_Customers.csv', index_col='CustomerID')
```

## Preparation des donn√©es

```python
x = data.copy()
```

```python
numerical_columns = data.select_dtypes(exclude = 'object').columns
categorical_columns = data.select_dtypes(include = 'object').columns
```

```python
scaler = StandardScaler()
x[numerical_columns] = scaler.fit_transform(x[numerical_columns])
```

```python
encoder = LabelEncoder()
x['gender_encoded'] = encoder.fit_transform(x[categorical_columns].values.ravel())
```

```python
x.drop(columns = 'Gender', inplace=True)
```

```python
x.head()
```

## Apply PCA on the features

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 0.92)
x_pca = pca.fit_transform(x)
```

```python
pca_explained_variance = pca.explained_variance_ratio_
```

```python
print(f' {'Principal Component'} : Explained Variance %')
for pc, var in enumerate(pca_explained_variance):
    print(f' PC_{str(pc+1):<22} : {var*100: .2f} %')
```

```python
k_range = range(1, 12)

inertia = list()

for k in k_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_pca)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(k_range, inertia, marker= 'o')
```

## We will use silhouette score

```python
from sklearn.metrics import silhouette_score

silhouette_range = range(2, 12)

scores = list()

for k in silhouette_range:
    kmeans = KMeans(n_clusters=k)
    labels = kmeans.fit_predict(x_pca)
    scores.append(silhouette_score(x_pca, labels))
```

```python
plt.plot(silhouette_range, scores, marker= 'o')
```

 ## we choose K= 7

```python
kmeans = KMeans(n_clusters=7)

labels = kmeans.fit_predict(x_pca)
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = labels)
plt.scatter(centroids[: , 0],  centroids[:, 1], c= 'red', marker='x')
```
