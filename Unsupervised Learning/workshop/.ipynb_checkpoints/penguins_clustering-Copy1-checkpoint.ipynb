```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
from sklearn.datasets import load_iris, load_diabetes
import numpy as np
```

```python
data = pd.read_csv('data/penguins.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.sex.value_counts()
```

```python
data.sex = data.sex.replace('.', None)
```

```python
data.sex
```

```python
data.isnull().sum()
```

```python
data.sex = data.sex.fillna(data['sex'].mode()[0])
```

```python
numerical_cols = data.select_dtypes(include=['float']).columns
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].mean())
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['sex'] = data['sex'].map( {
        'MALE' : 0,
        'FEMALE' : 1
     })
```

```python
x = data.copy()
x[numerical_cols] = StandardScaler().fit_transform(x[numerical_cols])
```

```python
x.head()
```

```python
from sklearn.metrics import silhouette_score

k_range = range(2, 11)
inertia = list()
scores = list()
for k in k_range:
    kmeans = KMeans(n_clusters = k, n_init=10)
    labels = kmeans.fit_predict(x)
    inertia.append(kmeans.inertia_)
    scores.append(silhouette_score(x, labels))
    
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(k_range, inertia, marker='o')
axes[1].plot(k_range, scores, marker='o')
```

```python
for k, silhouette in enumerate(scores):
    print(f'silhouette for k={k+2} is : {silhouette}')
```

```python
kmeans = KMeans(n_clusters = 7, n_init=10)
labels = kmeans.fit_predict(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components= 0.90)
x_pca = pca.fit_transform(x)
pca.explained_variance_ratio_
```

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=labels)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
min_samples = 5  # 2*number_of_features
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(x)
distances, indices = neighbors_fit.kneighbors(x)
```

```python
distances = np.sort(distances[:, -1])
plt.figure(figsize=(8,4))
plt.plot(distances, marker='o')
plt.xlabel('Points sorted by distance')
plt.ylabel(f'{min_samples}-th nearest neighbor distance')
plt.title('k-distance graph to find eps')
plt.show()
```

```python
from kneed import KneeLocator

# Detect the knee automatically
kneedle = KneeLocator(range(len(distances)), distances, curve='convex', direction='increasing')
eps = distances[kneedle.knee]
print(f"Recommended eps: {eps:.3f}")
```
