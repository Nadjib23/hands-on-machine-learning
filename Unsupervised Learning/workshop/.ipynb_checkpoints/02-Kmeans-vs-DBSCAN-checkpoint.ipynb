```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans
from sklearn.datasets import make_moons, make_blobs, make_circles
from sklearn.preprocessing import StandardScaler
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Create subplots for different types of distributions
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle('DBSCAN vs K-means Clustering Comparison', fontsize=16, fontweight='bold')

# Dataset 1: Spherical clusters (Blobs)
x_blobs, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=42)
x_blobs = StandardScaler().fit_transform(x_blobs)

# Dataset 2: non-spherical cluster (Moons)
x_moons, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
x_moons = StandardScaler().fit_transform(x_moons)

# Dataset 3: Circles
x_circles, _ = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)
x_circles = StandardScaler().fit_transform(x_circles)

datasets = [
    (x_blobs, "Blob-shaped Clusters"),
    (x_moons, "Moon-shaped Clusters"),
    (x_circles, "Circular Clusters")
]

for idx, (X, title) in enumerate(datasets):
    # Original data
    axes[idx, 0].scatter(X[:, 0], X[:, 1], s=30, alpha=0.7)
    axes[idx, 0].set_title(f'{title}\n(Original Data)')
    axes[idx, 0].set_xlabel('Feature 1')
    axes[idx, 0].set_ylabel('Feature 2')
    
    # K-means clustering
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    kmeans_labels = kmeans.fit_predict(X)
    
    axes[idx, 1].scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=30, 
                         cmap='viridis', alpha=0.7)
    axes[idx, 1].scatter(kmeans.cluster_centers_[:, 0], 
                         kmeans.cluster_centers_[:, 1],
                         c='red', marker='X', s=200, 
                         edgecolors='black', linewidths=2, label='Centroids')
    axes[idx, 1].set_title(f'K-means (k=2)\nClusters: {len(set(kmeans_labels))}')
    axes[idx, 1].set_xlabel('Feature 1')
    axes[idx, 1].legend()
    
    # DBSCAN clustering
    # Adjust eps based on dataset
    eps_values = [0.3, 0.3, 0.4]  
    dbscan = DBSCAN(eps=eps_values[idx], min_samples=5)
    dbscan_labels = dbscan.fit_predict(X)
    
    # Separate noise points
    mask_noise = dbscan_labels == -1
    mask_clusters = dbscan_labels != -1
    
    axes[idx, 2].scatter(X[mask_clusters, 0], X[mask_clusters, 1], 
                         c=dbscan_labels[mask_clusters], s=30, 
                         cmap='viridis', alpha=0.7)
    axes[idx, 2].scatter(X[mask_noise, 0], X[mask_noise, 1], 
                         c='red', marker='x', s=50, alpha=0.5, label='Noise')
    
    n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
    n_noise = list(dbscan_labels).count(-1)
    axes[idx, 2].set_title(f'DBSCAN (eps={eps_values[idx]})\nClusters: {n_clusters}, Noise: {n_noise}')
    axes[idx, 2].set_xlabel('Feature 1')
    axes[idx, 2].legend()

plt.tight_layout()
plt.savefig('clustering_comparison.png', dpi=300, bbox_inches='tight')
print("Visualization saved as 'clustering_comparison.png'")
```

```python

# Example 1: K-means
print("\n1. K-MEANS CLUSTERING:")
print("-" * 40)
X_example, _ = make_blobs(n_samples=100, centers=3, random_state=42)
X_example = StandardScaler().fit_transform(X_example)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_example)

print(f"Number of clusters: {kmeans.n_clusters}")
print(f"Cluster centers shape: {kmeans.cluster_centers_.shape}")
print(f"Inertia (sum of squared distances): {kmeans.inertia_:.2f}")
print(f"Sample labels: {labels[:10]}")
```

```python
# Example 2: DBSCAN
print("\n2. DBSCAN CLUSTERING:")
print("-" * 40)
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X_example)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"eps (radius): 0.3")
print(f"min_samples: 5")
print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")
print(f"Sample labels: {labels[:10]}")
print(f"Core sample indices shape: {dbscan.core_sample_indices_.shape}")
```
